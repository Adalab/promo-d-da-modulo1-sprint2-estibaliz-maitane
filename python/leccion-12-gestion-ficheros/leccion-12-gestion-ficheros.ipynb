### Creación de carpetas y movimientos entre ellas
---------------------------------- PRIMERA FUNCIÓN -----------------------   

Esta función debe incluir el código que:    

1. Nos muestre en que carpeta estamos trabajando.   

2. Cree una carpeta que se llame "aprendiendo-ficheros". ⚠️ Tened en cuenta que si la carpeta ya existe no la podemos crear, nos devolverá un error. Incluye en la función un programa que evite que nos de un error si la carpeta ya existe.

3. Cree otra carpeta que se llame "datos" dentro de la carpeta "aprendiendo-ficheros". En esta carpeta "datos" guardaremos el fichero "saludo.txt" que os habéis descargado.

4. Cambiad el directorio de trabajo a la carpeta "datos". Antes de seguir chequead que estáis trabajando en la carpeta "datos".  

5. Cambiad el nombre de la carpeta creada en el punto 2 a "primera-toma-contacto"  

    💡 **Pistas para resolver este ejercicio** 💡   
        Esta función debe incluir el código que:  
    1. La función tendrá que recibir 3 parámetros:   
        - El nombre del nombre de la primera carpeta  
        - El nombre de la carpeta segunda carpeta  
        - El nombre con el que queramos cambiar el nombre de la primera carpeta creada       
    2. Para saber si las carpetas ya existen tendréis que usar `listdir` (recordad que nos devuelve una lista de ficheros y carpetas).
    3. Para poder controlar los errores tendremos que usar un if loop, que si el fichero existe, nos devuelva un mensaje de que el fichero ya existe. En caso de que no exista, los deberéis crear y que la función nos muestre un mensaje de que se ha creado.
    4. Tendréis que ir cambiando de directorio para poder crear las carpetas y cambiar sus nombres.  

```python
# Definimos la función
def gestion_carpetas (carpeta, subcarpeta, camb_nombre):
    print(os.getcwd())
    lista_dir = os.listdir()
    if carpeta not in lista_dir:
        os.mkdir(carpeta)
        print(f'Se ha creado la carpeta "{carpeta}"')
    else:
        print(f'La carpeta {carpeta} ya existe')
    lista_subdir = os.listdir(f'./{carpeta}')
    os.chdir(f'./{carpeta}')
    if subcarpeta not in lista_subdir:
        os.mkdir(subcarpeta)
        print(f'Se ha creado la carpeta "{subcarpeta}"')
    else:
        print(f'La carpeta {subcarpeta} ya existe')
    os.chdir(f'./{subcarpeta}')
    print(os.getcwd())
    os.chdir('../../')
    os.rename(carpeta, camb_nombre)
    print('Se ha realizado correctamente el cambio de nombre')
    return 'La gestión de carpetas solicitada se ha realizado con éxito' 
```

```python
# Ejecutamos la función con estos argumentos
nombre_carpeta = "aprendiendo-ficheros"
nombre_subcarpeta = "datos"
camb_nombre = "primera-toma-contacto"
```

```python
gestion_carpetas(nombre_carpeta, nombre_subcarpeta, camb_nombre)
```

>📌 Antes de empezar, recordad descargaros el fichero `saludo.txt` y guardarlo en el repo en el que estáis trabajando, dentro de una carpeta que se llame "datos".  

1. Lea el fichero que se llame "saludo.txt y muestre su contenido completo.  

1. Muestra la línea 4 del fichero  

    💡 **Pistas para resolver este ejercicio** 💡  
    1. Antes de empezar, tendréis que saber cuál es vuestro directorio de trabajo.  
    2. Tened en cuenta en que carpeta estáis. Si vuestro directorio de trabajo no es "datos" tendréis que cambiarlo o poner la ruta relativa a la carpeta "datos".  
    3. Usar `if... else` para evitar que se nos pare el código.  
    4. Para cambiar el fichero podréis usar el comando `input` para preguntar el usuario donde está el fichero y que se pueda usar la ruta relativa o absoluta.  

```python
# Definimos la función
def abrir_txt (ruta_carpeta, nombre_archivo):
    print(os.getcwd())
    nombre_archivo = nombre_archivo.replace('.txt', '')
    nombre_txt = nombre_archivo + '.txt'
    ruta_archivo = ruta_carpeta + '/' + nombre_txt
    archivos_existentes = os.listdir(ruta_carpeta)
    if nombre_txt in archivos_existentes:
        texto_txt = open(ruta_archivo, encoding = 'utf-8') 
        lista_txt = texto_txt.readlines()
        print(f'El contenido del archivo completo es: {lista_txt}')
        texto_txt.close() 
        return f'La línea 4 del fichero es: {lista_txt[3]}'
    else:
        print('El archivo indicado no se encuentra en el directorio especificado')
```

```python
# Ejecutamos la función:
ubic_carpeta = 'primera-toma-contacto/datos'
nomb_archivo = 'saludo.txt'
abrir_txt(ubic_carpeta, nomb_archivo)
```

```python
def peliculas_xml(archivo_xml, ubic_carpeta):
    archivo_xml = archivo_xml.replace('.xml', '')
    ubic_carpeta = ubic_carpeta + '/' + archivo_xml + '.xml'
    tree = ET.parse(ubic_carpeta) 
    root = tree.getroot() 
    # Para facilitar la lectura separamos la obtención de las etiquetas y atributos de los textos de descripción y año
    # obtenemos las etiquetas y atributos
    print(f'Etiqueta raíz: {root.tag}')
    for hijo in root: 
        print(f'Etiqueta hijo: {hijo.tag}')
        print(f'Atributos hijo: {hijo.attrib}')
        for nieto in hijo:
            print(f'..Etiqueta nieto: {nieto.tag}')
            print(f'..Atributos nieto: {nieto.attrib}')
            for biznieto in nieto:
                print(f'....Etiqueta biznieto: {biznieto.tag}')
                print(f'....Atributos biznieto: {biznieto.attrib}')
                for tataranieto in biznieto:
                    print(f'......Etiqueta tataranieto: {tataranieto.tag}')
                    if tataranieto.attrib != {}:
                        print(f'     * Atributos tataranieto: {tataranieto.attrib}')
    
    # Para obtener las descripciones de las películas y el año de estreno:
    for hijo in root:
        for nieto in hijo:
            for biznieto in nieto: 
                print('Descripción película:' + biznieto.find('description').text)
                print('Año estreno:' + biznieto.find('year').text)
    
    
```
